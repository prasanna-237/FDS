```python
import numpy as np
import warnings
warnings.filterwarnings('ignore')
array=np.random.randint(1,100,16) 
array
```

```python
array.mean()
```

```python
np.percentile(array,25)
```

```python
np.percentile(array,50)
```

```python
np.percentile(array,75)
```

```python
np.percentile(array,100)
```

```python
#outliers detection
def outDetection(array):
    sorted(array)
    Q1,Q3=np.percentile(array,[25,75])
    IQR=Q3-Q1
    lr=Q1-(1.5*IQR)
    ur=Q3+(1.5*IQR)
    return lr,ur
lr,ur=outDetection(array)
lr,ur
```

```python
import seaborn as sns
%matplotlib inline
sns.displot(array)
```

```python
sns.distplot(array)
```

```python
new_array=array[(array>lr) & (array<ur)]
new_array
```

```python
sns.displot(new_array)
```

```python
lr1,ur1=outDetection(new_array)
lr1,ur1
```

```python
final_array=new_array[(new_array>lr1) & (new_array<ur1)]
final_array
```

```python
sns.distplot(final_array)
```

```python
#EX.NO :6  Random Sampling and Sampling Distribution
#DATA  : 10.09.2024

#NAME : PRASANNA KUMAR M
#ROLL NO : 230701237
#DEPARTMENT : B.E COMPUTER SCIENCE AND ENGINEERING - D
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
population_mean = 50
population_std = 10
population_size = 100000
population = np.random.normal(population_mean, population_std, population_size)
```

```python
sample_sizes = [30, 50, 100]
num_samples = 1000
```

```python
sample_means = {}
for size in sample_sizes:
   sample_means[size] = []
   for _ in range(num_samples):
      sample = np.random.choice(population, size=size, replace=False)
      sample_means[size].append(np.mean(sample))
```

```python
plt.figure(figsize=(12, 8))
```

```python
for i, size in enumerate(sample_sizes):
   plt.subplot(len(sample_sizes), 1, i+1)
   plt.hist(sample_means[size], bins=30, alpha=0.7, label=f'Sample Size {size}')
   plt.axvline(np.mean(population), color='red', linestyle= 'dashed', linewidth=1.5,
label= 'Population Mean')
   plt.title(f'Sampling Distribution(Sample Size {size})')
   plt.xlabel('Sample mean')
   plt.ylabel('Frequency')
   plt.legend()
plt.tight_layout()
plt.show()
```

```python
#EX.NO :7  Z-Test
#DATA  : 10.09.2024

#NAME : PRASANNA KUMAR M
#ROLL NO : 230701237
#DEPARTMENT : B.E COMPUTER SCIENCE AND ENGINEERING - D
```

```python
import numpy as np
import scipy.stats as stats
```

```python
sample_data = np.array([152, 148, 151, 149, 147, 153, 150, 148, 152,
149,151, 150, 149, 152, 151, 148, 150, 152, 149, 150,148, 153, 151,
150, 149, 152, 148, 151, 150, 153])
```

```python
population_mean = 150
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
```

```python
n = len(sample_data)
z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))
```

```python
# Assuming sample_mean, z_statistic, and p_value have already been calculated:
print(f"Sample Mean: {sample_mean:.2f}\n")
print(f"Z-Statistic: {z_statistic:.4f}\n")
print(f"P-Value: {p_value:.4f}\n")

# Significance level
alpha = 0.05

# Decision based on p-value
if p_value < alpha:
    print("Reject the null hypothesis: The average weight is significantly different from 150 grams.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in average weight from 150 grams.")
```

```python
import numpy as np
import scipy.stats as stats
np.random.seed(42)
sample_size = 25
sample_data = np.random.normal(loc=102, scale=15, size=sample_size)
```

```python
population_mean = 100
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
```

```python
n = len(sample_data)
t_statistic, p_value = stats.ttest_1samp(sample_data,population_mean)
```

```python
# Assuming sample_mean, t_statistic, and p_value have already been calculated:
print(f"Sample Mean: {sample_mean:.2f}\n")
print(f"T-Statistic: {t_statistic:.4f}\n")
print(f"P-Value: {p_value:.4f}\n")

# Significance level
alpha = 0.05

# Decision based on p-value
if p_value < alpha:
    print("Reject the null hypothesis: The average IQ score is significantly different from 100.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in average IQ score from 100.")
```

```python
#EX.NO :9  Annova TEST
#DATA  : 08.10.2024

#NAME : PRASANNA KUMAR M
#ROLL NO : 230701237
#DEPARTMENT : B.E COMPUTER SCIENCE AND ENGINEERING - D
```

```python
import numpy as np
import scipy.stats as stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

np.random.seed(42)
n_plants = 25
```

```python
growth_A = np.random.normal(loc=10, scale=2, size=n_plants)
growth_B = np.random.normal(loc=12, scale=3, size=n_plants)
growth_C = np.random.normal(loc=15, scale=2.5, size=n_plants)
```

```python
all_data = np.concatenate([growth_A, growth_B, growth_C])
```

```python
treatment_labels = ['A'] * n_plants + ['B'] * n_plants + ['C'] * n_plants
f_statistic, p_value = stats.f_oneway(growth_A, growth_B, growth_C)
```

```python
mean_A = np.mean(growth_A)
mean_B = np.mean(growth_B)
mean_C = np.mean(growth_C)
print(f"Treatment A Mean Growth: {mean_A:.4f}")
print(f"Treatment B Mean Growth: {mean_B:.4f}")
print(f"Treatment C Mean Growth: {mean_C:.4f}")
print(f"F-Statistic: {f_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in mean growth rates among the three treatments.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in mean growth rates among the three treatments.")


if p_value < alpha:
  
    tukey_results = pairwise_tukeyhsd(all_data, treatment_labels, alpha=0.05)
    
    print("\nTukey's HSD Post-hoc Test:")
    print(tukey_results)
```
